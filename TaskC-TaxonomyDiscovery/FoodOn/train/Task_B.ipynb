```python
from transformers import pipeline
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load the feature extraction pipeline with RecipeBERT
extractor = pipeline(
    'feature-extraction', model='alexdseo/RecipeBERT', framework='pt'
)

# Function to get mean-pooled embedding for a given text
def get_embedding(text: str):
    outputs = extractor(text)[0]  # shape: [1, sequence_length, hidden_dim]
    return np.mean(outputs, axis=1)  # [1, hidden_dim]

# Define two food terms
term1 = "Hawaiian Pizza"
term2 = "Margherita Pizza"

# Get embeddings
vec1 = get_embedding(term1)
vec2 = get_embedding(term2)

# Compute cosine similarity
similarity = cosine_similarity(vec1, vec2)[0][0]

print(f"Cosine similarity between '{term1}' and '{term2}': {similarity:.4f}")
```
