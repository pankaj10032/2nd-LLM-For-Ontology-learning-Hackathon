```python
def split_list_into_chunks(data_list, chunk_size=100):
  """
  Splits a list into smaller lists of a specified size.

  Args:
    data_list: The list to be split.
    chunk_size: The desired size of each chunk.

  Returns:
    A list of lists, where each inner list is a chunk.
  """
  if not data_list:
    return []
  
  # Use a list comprehension to create the chunks
  # The range function steps by chunk_size to get the start of each new chunk
  chunks = [data_list[i:i + chunk_size] for i in range(0, len(data_list), chunk_size)]
  return chunks
```
