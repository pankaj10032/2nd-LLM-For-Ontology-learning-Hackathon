```python
import json
```

```python
def load_json(data_path):
    with open(data_path, "r") as file_1:
        data = json.load(file_1)
    return data
```

```python
train_data = load_json("GO/train/go_train_re_pairs.json")
train_data
```

```python
len(train_data)
```

```python
from collections import defaultdict
```

```python
grouped = defaultdict(list)
for entry in train_data:
    head = entry.get('head')
    if head:
        grouped[head].append(entry)

# Convert to list of lists
result = list(grouped.values())

# Output the result
for group in result:
    print(group)
```
