```python
generation_config = {"response_mime_type": "application/json"}

response = client.models.generate_content(
        contents= prompt,
        config=generation_config,
        model=model
    )
```

```python
response_2 = response.text
response_2
```

```python
ans = json.loads(response_2)
```

```python
ans
```

```python
clustered_terms_domain = {}
for term, domains in ans.items():
    for domain in domains:
        if domain not in clustered_terms_domain.keys():
            clustered_terms_domain[domain]= []
        clustered_terms_domain[domain].append(term)
```
